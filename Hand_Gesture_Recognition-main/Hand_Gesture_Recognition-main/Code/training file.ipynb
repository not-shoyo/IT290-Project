```python
#initializing a list of paths for images
imagepaths = []
```

```python
#extracting dataset
import os
print(os.getcwd())
path_diff = len(os.getcwd().split(os.sep))
print(path_diff)
if os.getcwd().split(os.sep)[-1] == "Code":
    os.chdir(os.pardir + '\\leapGestRecog')
print(os.getcwd())
# for dirname, _, filenames in os.walk(r'C:\Users\hp\Desktop\archive\leapGestRecog'):
# for dirname, _, filenames in os.walk(r'leapGestRecog'):
for dirname, _, filenames in os.walk(os.getcwd()):
    for filename in filenames:
        path = os.path.join(dirname, filename)
        # print(os.path.join(dirname,filename))
        if path.endswith("png"):
            imagepaths.append(path)
            
print(len(imagepaths))
```

```python
# creating training set and labels
# will have to change the number path_list[idx][1] for each user depending on previous blocks path output

x_list = []                                         # x for image data
y_list = []                                         # y for the labels

# print(path.split(os.sep))


#load the images into x by doing necessary conversions and resizing the image
#resizing is done to reduce the size of image to increase the speed of training

for path in imagepaths[:19999]:
    img = cv2.imread(path)                          # get images
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)     # gray-scale images

    img = cv2.resize(img, (128,128))                # resizing to increase training speed
    x_list.append(img)                              # adding image to x_list
    path_list = path.split(os.sep)

    x_label=path_list[path_diff + 1][1]             # getting the label to add to label list (taking into)
    y_list.append(x_label)                          # adding to label list
    
    # print(path_list)
    print(x_label)

x_list=np.array(x_list)
print(len(x_list))
print(len(y_list))
print(x_list.shape)
```

```python
# splitting the dataset as training and test dataset
# we train the model with 10 gestures but actually use only 8 of them
# used random_state initialization to see effect on model but no discernable improvement.

x_train, x_test, y_train, y_test = train_test_split(x_list, y_list, test_size = 0.3, train_size = 0.7, random_state = 0)  # 70-30 split into training and testing data


# since machine learning models require numerical values for training, it is necessary to "categorize" them with numerical values
# each row contains 0s with only one 1 value corresponding to the label to be assigned to it.
# the matrix created is called a class matrix.

y_train = np_utils.to_categorical(y_train, num_classes = 10)                                                              # creating class matrix for training data
y_test = np_utils.to_categorical(y_test, num_classes = 10)                                                                # creating class matrix for testing data

print("Shape after one-hot encoding: ", y_train.shape)                                                                    # crosschecking shape after one-hot encoding (expected n number of training entries and row of 10 columns each with only 0s and one 1)
print(len(x_train),len(x_test),len(y_train),len(y_test))                                                                  # seeing how many training & testing - inputs & labels
print(x_train.shape)                                                                                                      # crosschecking shape of training data (expected n number of training entries and each entry has size 128 * 128)
```

```python
# creating CNN model

from tensorflow import keras as keras
from keras.models import Sequential
from keras.layers.convolutional import Conv2D,MaxPooling2D
from keras.layers import Dense,Dropout,Flatten


# Sequential models have one input and one output, like how an input is one image and it has a single output label.
# Since CNN models are a class of Deep Learning / Deep Neural Networks, this model will need multiple layers,
# which have to be added along side their convolutional layers (kernels used to extract features from images) and maxpooling (returning the maximum value from an area in the image by ampplitude)
# each layer also has an activation function named (like signmoid etc), we use 'relu' which stands for rectified linear activation function. 
# working similarly to a normal linear activation function, where 0 is returned for any non positive value else the value itself is returned

model = Sequential()
model.add(Conv2D(32, (5,5), activation = 'relu', input_shape=(128,128,1)))    # first layer in keras Sequential model will reuire us to provide input size, thereafter, the model will infer on its own.
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(64, (3, 3), activation='relu')) 
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu')) 
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu')) 
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu')) 
model.add(MaxPooling2D((2, 2)))


# after model is defined, adding extra final layers 
# for making it compatible with our data and useful for statistical analysis later.

model.add(Flatten())                                                          # making single dimentional to be able to train on data
model.add(Dropout(0.5))                                                       # necessary for regularization in the model. (dropping random nueral nodes ie temporarily suspending their involvement on the forward and backward passes)
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))                                    # making all 10 output labels into probability distributions
```

```python
# compiling the model to prepare for training. 
# we provide parameters like type of optimizer and loss (the goal is to minimize loss)

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# train and test input data are reshaped since the model.fit() requires certain parameters of definite shape

print(x_train.shape)                                                  # noting shapes before reshaping
print(x_test.shape)                                                   # noting shapes before reshaping

x_train=x_train.reshape(x_train.shape[0],128,128,1)                   # reshaping training data
x_test=x_test.reshape(x_test.shape[0],128,128,1)                      # reshaping training data

print(x_train.shape)                                                  # printing after reshaping 
print(x_test.shape)                                                   # printing after reshaping
```

```python
# training the model

model.fit(x_train, y_train, epochs=10, batch_size=2, verbose=2, validation_data=(x_test, y_test))
print('Model training complete')
```

```python
# saving the trained model

os.chdir('../Code/')
print(os.getcwd())
model.save("handgestures.h5")
print("Model saved")
```

```python
# Testing the model
# calculate loss and accuracy on test data

tLoss, tAccuracy = model.evaluate(x_test, y_test)

print('Test accuracy: {:2.2f}%'.format(tAccuracy*100))
```

```python
# Function to plot images and labels for validation
def validate_gestures(predictions_array, true_label_array, img_array):
  # Array for pretty printing and then figure size
  #class_names = ["down", "palm", "l", "fist", "fist_moved", "thumb", "index", "ok", "palm_moved", "c"] 
  class_names = ["down", "palm", "L", "fist", "fist_moved", "thumb", "index", "ok", "palm_moved", "C"] 
  plt.figure(figsize=(15,5))
  
  for i in range(1, 10):
    # assigning variables
    prediction = predictions_array[i]
    true_label = true_label_array[i]
    img = img_array[i]
    img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
    
    # structured plots
    plt.subplot(3,3,i)
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img, cmap=plt.cm.binary)

    predicted_label = np.argmax(prediction)                   # Get index of the predicted label from prediction

    if predicted_label == true_label:                         # Change color of title based correctness of prediction
      color = 'blue'
    else:
      color = 'red'

    plt.xlabel("Predicted: {} {:2.0f}% (Actual: {})".format(class_names[predicted_label], 100*np.max(prediction), class_names[true_label]), color=color)
plt.show()
```

```python
# Plot testing based on predictions and their actual values
validate_gestures(prediction, y_test_new, x_test)
```

```python
#Transform predictions into 1D array 
y_pred = np.argmax(prediction, axis=1)
```

```python
#Create a Confusion Matrix for Evaluation
# H = Horizontal
# V = Vertical
table = pd.DataFrame(confusion_matrix(y_test_new, y_pred), 
             columns=["Predicted Palm", "Predicted L", "Predicted Fist", "Predicted Fist (V)", "Predicted Thumb", "Predicted Index", "Predicted OK", "Predicted Palm Moved", "Predicted C","Predicted Down"],
             index=["Actual Palm", "Actual L", "Actual Fist", "Actual Fist (V)", "Actual Thumb", "Actual Index", "Actual OK", "Actual Palm Moved", "Actual C","Actual Down"])
table.drop(index=["Actual Fist (V)","Actual OK"], columns=["Predicted Fist (V)","Predicted OK"])
```
